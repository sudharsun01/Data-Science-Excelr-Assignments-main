```python
user_book = book.pivot_table(index='User.ID',
                       columns='Book.Title',
                       values='Book.Rating')
```

```python
user_book
```

```python
user_book.fillna(0, inplace=True)
```

```python
user_book
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_book, metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids
user_sim_df.index = book['User.ID'].unique()
user_sim_df.columns = book['User.ID'].unique()
```

```python
def recommend(ID):
  simID = user_sim_df.loc[ID].idxmax()
  simID_book = book[book['User.ID'] == simID]['Book.Title']
  custID_books = book[book['User.ID'] == ID]['Book.Title']
  return set(simID_book) - set(custID_books)
```
